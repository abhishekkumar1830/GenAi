**Vector**
  - Ai models does not understand text,he understand numbers only
  - A numerical representation of text is called vectors

**Embedding**
   - Any numbers is called vectors but the meaning of numbers is embedding
   - it is a process of converting text in numbers with some meaning

- Why vectors?
   - Because vector gives direction if the vector has same direction means text has similiar meaining,if different directions both text has different          meaning

**Cosine similarity**
   - It measure how similar two vectors are by checking the angle between them
   - basically it checks how similar the meaining of two text are
   - *1* is exactly same meaning
   - *0* is no relation
   - *-1* is opposite meaning

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#two square brackets bcz machine larning functions expect vectors as in 2d array
v1=np.array([[1,2]])
v2=np.array([[2,4]])
v3=np.array([[-1,-2]])
```

```python
#same direction and very similar
cosine_similarity(v1,v2)
```

```python
#opposite direction
cosine_similarity(v1,v3)
```

```python
from sentence_transformers import SentenceTransformer
sentence=["cat is sleep on sofa",
          "virat kohli has 37 years of age",
          "the capital of India is Delhi",
          "Trump is the current President of USA"]
```

```python
#here i am taking ai brain name: all-MiniLM-L6-V2, its free protocol model
#MiniLM → tiny focused brain, only understands meaning and outputs vectors, doesn’t generate text
model=SentenceTransformer("all-MiniLM-L6-V2")
model
```

```python
#encode is here to give sentences to model,now model convert text-numbers-embedding(meaning to numebers) then encode returns vector
embeddings=model.encode(sentence)
len(embeddings)
```
